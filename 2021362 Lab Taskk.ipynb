```python
from collections import deque

# define a function for BFS
def bfs(graph, start):
    visited = set() # set to keep track of visited nodes
    queue = deque([start]) # initialize the queue with the start node
    while queue: # loop until queue is empty
        vertex = queue.popleft() # get the first node from the queue
        if vertex not in visited:
            visited.add(vertex) # mark the current node as visited
            print(vertex) # print the visited node
            queue.extend(graph[vertex] - visited) # add the unvisited neighbors to the queue
    return visited # return the set of visited nodes
```

```python
# example usage
graph = {
    'A': {'B', 'C'},
    'B': {'A', 'D', 'E'},
    'C': {'A', 'F'},
    'D': {'B'},
    'E': {'B', 'F'},
    'F': {'C', 'E'}
}
bfs(graph, 'A') # start BFS from node 'A'
```

```python
# define a function for DFS
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set() # set to keep track of visited nodes
    visited.add(start) # mark the current node as visited
    print(start) # print the visited node
    for next_node in graph[start] - visited: # visit unvisited neighbors recursively
        dfs(graph, next_node, visited)
    return visited # return the set of visited nodes
```

```python
# example usage
graph = {
    'A': {'B', 'C'},
    'B': {'A', 'D', 'E'},
    'C': {'A', 'F'},
    'D': {'B'},
    'E': {'B', 'F'},
    'F': {'C', 'E'}
}
dfs(graph, 'A')
```

```python
import random

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def insert(root, val):
    if root is None:
        return Node(val)
    if val < root.val:
        root.left = insert(root.left, val)
    else:
        root.right = insert(root.right, val)
    return root

def build_bst(nums):
    root = None
    for num in nums:
        root = insert(root, num)
    return root

# Generate random and unique numbers for each set
set1 = random.sample(range(1, 1001), 100)
set2 = random.sample(range(1, 40001), 1000)
set3 = random.sample(range(1, 80001), 2000)
set4 = random.sample(range(1, 200001), 5000)
set5 = random.sample(range(1, 1000001), 10000)

# Build binary search trees for each set
bst1 = build_bst(set1)
bst2 = build_bst(set2)
bst3 = build_bst(set3)
bst4 = build_bst(set4)
bst5 = build_bst(set5)
```

```python
from collections import deque

def bfs(root):
    if root is None:
        return []
    queue = deque([root])
    result = []
    while queue:
        node = queue.popleft()
        result.append(node.val)
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    return result

# Perform BFS on each of the five binary search trees
bfs1 = bfs(bst1)
bfs2 = bfs(bst2)
bfs3 = bfs(bst3)
bfs4 = bfs(bst4)
bfs5 = bfs(bst5)

print(bfs1)
print(bfs2)
print(bfs3)
print(bfs4)
print(bfs5)
```

```python
from queue import PriorityQueue

class Node:
    def __init__(self, state, parent=None, cost=0):
        self.state = state
        self.parent = parent
        self.cost = cost

    def __lt__(self, other):
        return self.cost < other.cost

def ucs(start_state, goal_state, actions_fn, cost_fn):
    start_node = Node(start_state)
    frontier = PriorityQueue()
    frontier.put(start_node)
    explored = set()

    while not frontier.empty():
        curr_node = frontier.get()
        curr_state = curr_node.state

        if curr_state == goal_state:
            path = []
            while curr_node:
                path.append(curr_node.state)
                curr_node = curr_node.parent
            return path[::-1]

        explored.add(curr_state)
        for action in actions_fn(curr_state):
            new_state = action
            new_cost = curr_node.cost + cost_fn(curr_state, action)
            new_node = Node(new_state, curr_node, new_cost)

            if new_state not in explored:
                frontier.put(new_node)
            else:
                for node in frontier.queue:
                    if node.state == new_state and node.cost > new_cost:
                        frontier.get(node)
                        frontier.put(new_node)

    return None
```

```python
def path_cost(path):
    total_cost=0 #store cost of whole path
    for (node,cost) in path: #iterate over path given
        total_cost+=cost
    return total_cost,path[-1][0] #return the total path cost and end node in path
```

```python
def UCS(graph,start,goal):
    visited=[] # tore nodes visited
    queue=[[]]# store paths   [[path1],[path2],[(node1),(node2)]]
    queue[0].append((start,0)) #start exploring from start in 1st path. cost=0 b/c we already there
    
    while queue: #while an unexplored path left
        queue.sort(key=path_cost) #sort queue paths ascending based on total_cost
        path=queue.pop(0) #get path with least cost and remove it from queue
        node=path[-1][0] #last node visited in this path
        if node in visited:
            continue #if node is already visisted i.e path ended move to other path
        visited.append(node) #mark node as visited
        if node==goal:
            return path,queue #if goal reached 
        else:
            adjacent_nodes = graph.get(node,[]) #get adjacent nodes from graph {node: [(adj,cost),(adj,cost)]}
            for (adj_node,cost) in adjacent_nodes: #traverse over all adjacent nodes
                new_path=path.copy()
                new_path.append((adj_node,cost)) #newpath = path till now + adjacent node
                queue.append(new_path)
```

```python
graph = {'S': [('A', 1), ('G', 12)], 
         'A': [('S', 1), ('B', 3), ('C', 1)], 
         'G': [('S', 12), ('C', 2), ('D', 3)], 
         'B': [('A', 3), ('D', 3)], 
         'C': [('A', 1), ('D', 1), ('G', 2)], 
         'D': [('C', 1), ('B', 3), ('G', 3)]}

solution,queue=UCS(graph,'S','G')
print("Path folowed: ",solution)
print("Path cost:",path_cost(solution)) 
```

```python
from queue import Queue

# Function to check if a position is safe for the queen
def is_safe(board, row, col):
    # Check left diagonal
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    # Check right diagonal
    for i, j in zip(range(row, -1, -1), range(col, len(board))):
        if board[i][j] == 1:
            return False
    # Check column
    for i in range(row):
        if board[i][col] == 1:
            return False
    return True

# Function to print the board
def print_board(board):
    for row in board:
        print(' '.join([str(elem) for elem in row]))

# Function to solve the n-queens problem using BFS
def n_queens_bfs(n):
    # Initialize an empty board
    board = [[0 for x in range(n)] for y in range(n)]
    # Initialize a queue with the first row of the board
    q = Queue()
    for i in range(n):
        q.put([(0, i)])
    # Iterate until a solution is found or the queue is empty
    while not q.empty():
        path = q.get()
        # If the path reaches the end, we have a solution
        if len(path) == n:
            for row, col in path:
                board[row][col] = 1
            return board
        # Otherwise, extend the path with safe positions in the next row
        row = len(path)
        for col in range(n):
            if is_safe(board, row, col):
                new_path = path + [(row, col)]
                q.put(new_path)
    # If the queue is empty, there is no solution
    return None

# Test the function with n = 4
board = n_queens_bfs(4)
if board:
    print_board(board)
else:
    print("No solution found")
```

```python
import heapq

# State class represents a state in the N-Queens problem
class State:
    def __init__(self, queens, cost):
        self.queens = queens  # queens is a list of column indices of queens
        self.cost = cost

    def __lt__(self, other):
        return self.cost < other.cost

    def __eq__(self, other):
        return self.queens == other.queens

    def __str__(self):
        return f"Queens: {self.queens}, Cost: {self.cost}"

# heuristic function that returns the number of conflicts for a given state
def num_conflicts(state):
    num = 0
    for i in range(len(state.queens)):
        for j in range(i+1, len(state.queens)):
            if state.queens[i] == state.queens[j] or \
               abs(state.queens[i] - state.queens[j]) == j - i:
                num += 1
    return num

# function that generates child states of a given state
def generate_children(state):
    children = []
    for i in range(len(state.queens)):
        for j in range(len(state.queens)):
            if j != state.queens[i]:
                new_queens = list(state.queens)
                new_queens[i] = j
                new_state = State(new_queens, state.cost + 1)
                heapq.heappush(children, new_state)
    return children

# UCS algorithm to solve the N-Queens problem
def ucs_n_queens(n):
    start_state = State([0]*n, 0)
    if num_conflicts(start_state) == 0:
        return start_state
    frontier = []
    heapq.heappush(frontier, start_state)
    explored = set()
    while frontier:
        state = heapq.heappop(frontier)
        if num_conflicts(state) == 0:
            return state
        explored.add(state)
        children = generate_children(state)
        for child in children:
            if child not in explored:
                heapq.heappush(frontier, child)

# example usage: solve the 8-Queens problem
solution = ucs_n_queens(8)
print(solution)
```
